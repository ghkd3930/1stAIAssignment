##인공지능 분야별 실제 적용 사례


* **[1] 언어 인공지능**

 * WT2 Plus 인공지능 통역기 이어폰

   * 40개 언어, 88개국 액센트를 지원하여 음성으로 입력하여 음성으로 출력되는  음성인식 통역 이어폰이다. 대개 이러한 음성 통역기의 경우 3단계 과정을 거쳐 음성 대 음성 변환 능력을 수행한다.

     1. 자동 음성 인식 소프트웨어가 말을 문자로 변환한다.
     2. 신경망 기계번역 기술을 사용해 이 문자를 다른 언어의 문자로 변환한다.
     3. 문자 음성 변조기가 변환된 다른 문자를 음성으로 출력한다.

    * 이렇게 번역의 전 과정을 인공지능으로 번역하는 것을 신경망 기계 번역이라고 일컫는다. 이는 원문 번역의 전 과정을 인공 신경망으로 처리하는 기계 번역을 뜻하며, 이 경우는 인공 신경망이 원문과 해당 번역문 사이의 관계를 학습하기 때문에 기존의 통계적 기계 번역보다 우수한 성능을 보인다.


* **[2] 음성 인공지능**

 * Google Nest mini

   * 음성을 인식하여 기능을 수행하는 형식의 음성처리 인공지능의 사례는 무척이나 많지만 현재 가장 많이 사용하고 있는 제품은 *인공지능 스피커*다. 그 중 가장 많이 알려진 제품인 google nest mini의 경우 미세먼지, 유튜브 연동, 날씨확인, 알람 기능, 주변장소 등의 기능이 있다. 인공지능 기반 음성인식 제품에 말을 하면 **음성 입력 및 인식 -> 자연어 처리 -> 인식 결과**의 단계를 걸친다. 위에 기술한 언어 인공지능의 처리 과정과 비슷하나 실제 출력 방법이 다르기 때문에 각 기능들에 대해서 아래 자세히 서술해본다.

     1. 음성 입력 과정을 거쳐 사람의 음성을 컴퓨터가 이해할 수 있도록 텍스트화 한다. - 이 기술을 STT(Speech To Text)라고 한다. 주변 소음을 제외한 말소리를 파악하고 각 발음과 단어를 인식한 후 핵심어와 연결 단어를 인식해 입력한 후 텍스트로 옮기는 작업을 진행한다.

     2. 해당 문장을 자연어 처리(Natural Language Processing)한다. - 이 때 자연어 처리란 컴퓨터가 자연어를 분석하여 이해하고 처리하는 기술을 의미한다. 이 과정에서는 자연어에 대한 형태소 분석, 구문 해석, 의미 분석, 화용 분석 등을 통해 컴퓨터가 문장에 담긴 의도를 파악하게 한다.

     3. 인식 결과를 출력한다. - 인식된 요청에 따라 가장 최적의 결과를 찾아내고, TTS(Text to Speech) 기술을 통해 사람의 말소리처럼 응답하며 명령어에 해당되는 기능을 수행한다. 그것은 번역일수도 있고, 스피커의 경우는 음악을 트는 것일수도 있다.
     
    * 위에 기술된 과정 중 자연어 처리가 바로 인공지능의 주요 분야다. 인공지능 기계가 사람의 언어를 얼마나 잘 파악하였는지를 알 수 있는 과정이기 때문이다.


* **[3] 이미지 인공지능**

 * CheXNeXt Algorithm

   * 2017년 11월 20일 국제 학술지 'PLOS medicine' 온라인판에 실린 'CheXNeXt'는 미국 스탠퍼드대 연구팀에서 개발한 인공지능 알고리즘이다. 흉부 X선 이미지를 판독해 12가지 이상의 질환을 알아낼 수 있는 인공지능으로, 촬영한 부위가 어디인지 알아내는 것 뿐만 아니라 14가지 병리학적 증상을 알아낼 수 있다고 한다. 방사선과 전문의가 판독한 결과와 해당 알고리즘이 판독한 결과를 비교했을 때 14가지 중 10가지는 방사선과 전문의보다 정확하게 판단했음이 밝혀졌다. 또한, 이미지를 읽고 해당 증상을 알아내는 데에 소요되는 시간이 인공지능이 월등히 빠르다는 강점이 있다.

   * 현재 이미지 인공지능은 의학 분야에서 많은 각광을 받고 있다. 지난 AI with 구글 2019 코리아에서 구글 AI 프로덕트 매니저인 릴리 펭은 부족한 의사들을 대신해 의료 분야에서 AI를 확대해야 한다고 강조했다. 현재 영상의학 분야에서 AI의 신뢰도는 97% 수준을 넘나들고 있으며 같은 조건에서 놓고 볼 시 실제 사람보다 더 뛰어나다는 평가도 나온다.

   * 개인적으로 본인은 의료의 경우는 인간의 목숨과 직접적으로 연결되어있는 분야이기 때문에 인공지능을 맹신하기에는 아직은 이르다는 생각을 하고 있다. 하지만 인공지능의 가장 큰 강점은 바로 인간이 따라갈 수 없는 속도이다. 기본적인 초기 진단을 인공지능이 내리도록 하고, 세부 진단은 의사가 직접 살펴보는 것만 해도 많은 시간을 감축시킬 수 있을 것이다.


* **[4] 자율주행 인공지능**

 * 세종시 자율주행버스

   * 2019년 7월. 중소벤처기업부에서 세종시를 규제자율특구로 지정하면서 세종시 자율주행버스의 운행이 시작되었다. 규제자율특구란 규제로 인해 시험이 불가능한 혁신기술을 제약 없이 테스트할 수 있는 지역을 말하는데 이를 통해 세종시는 행복도시와 조치원 세종테크노파크 지역을 세종 자율주행 실증 규제자유특구로 지정하면서 세종시는 자율주행특구가 되어서 실제 운행에 들어갔다.

   * 2019년 10월 시범 주행한 자율주행버스는 현재 '레벨 3' 수준에 이르렀으며, 2021년에는 레벨 4 수준의 기술을 시민들에게 선보인다는 구상이다. 이 때 말하는 레벨이란 ADAS(Advanced Driver Assistance Systems)로 단계에 따른 설명은 아래와 같다.

     1. 운전자 지원 단계 : ADAS가 주행 환경을 인식해 조향 또는 가감속을 조작하나 나머지는 운전자가 조작하는 단계

     2. 부분 자동화 단계 : ADAS가 주행 환경을 인식해 조향과 가감속을 통합적으로 조작하나(적어도 2가지 기능 완전 자동화) 나머지는 운전자가 조작하는 단계.

     3. 조건부 자동화 단계 : 자율주행 시스템이 자동차를 감시하고 안전 관련 핵심 기능을 통제하나 때때로 운전자 개입을 요구하는 단계.

     4. 고도 자동화 단계 : 운전자가 자동차의 요청에 적절히 응하지 않더라도 대부분의 경우 자율주행 시스템 스스로 자동차를 통제하는 단계.

     5. 완전 자동화 단계 : 자율주행 시스템이 모든 도로와 조건에서 자동차를 조작하는 단계.

   * 이러한 자율주행 자동차를 위해 필요한 것은 인공지능과 센서이다. 자율주행 시스템의 센서는 자차의 위치 인식을 위한 센서와 주행환경에 존재하는 다양한 물체 인식을 위한 센서로 구성된다.자차의 정확한 위치정보가 획득된다면 자율주행지도를 참조하여 실세계 좌표계에서 전역 경로계획(Global Path Planning)을 세운다. 전역 경로계획으로 자율주행하면서 자차의 주행 경로 주변에서 이동하거나 정지된 물체를 인식하여 지역 경로계획(Local Path Planning)까지 수립하고, 샤시 시스템의 종/횡방향 제어를 수행함으로써 자율주행이 실행된다.

   * 어드벤스드 크루즈 컨트롤, 차선유지보조 시스템, 긴급 제동 보조 시스템(AEB) 등은 선행 차량 및 보행자와의 충돌이 예상될 경우 차량을 스스로 제동시켜 사고 위험을 경감시키는 기능으로, 이것에서 알고리즘에 학습기능을 부여하여 차량 전방에 탑재된 카메라에 인공지능의 머신러닝 기법이 사용된다. 현재 자율주행에 사용된 인공지능은 12개의 카메라를 비롯해 레이더, 초음파 센서 등 다양한 센서들을 통해 입력되는 정보들을 결합하여 주변 환경과 상황을 인식 및 인지하는데 사용하고 있다.
